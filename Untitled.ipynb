```python
m_list=[0, 1, 5, 10, 15, 20, 25, 35, 50]
error_list=[]
b=0

for each_m in m_list: 
    y=each_m*X+b
#     print(y.shape)
    total_error=(np.square(y.flatten()-y_true)).mean()
#     m_list.append(m)
    error_list.append(total_error)

    fig, ax=plt.subplots(1, 2)

    ax[0].scatter(X, y_true)
    ax[0].plot(X, y, color='r')

    ax[1].plot([n+1 for n in list(range(0, len(error_list)))], error_list)
    ax[1].set_xlim(0, 10)
    ax[1].set_ylim(0, 5000)
    
    fig.savefig(f'images/image_{each_m}')
```
